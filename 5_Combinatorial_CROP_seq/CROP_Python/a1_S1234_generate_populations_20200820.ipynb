# II. Add fitness to metadata of cells

**1. No fitness data for pop1.cells at Day0**

    perturbed_1: perturbations at sgRNA level.
    perturbed_1_Ab.ordered: same with perturbed_1, except ordering two perturbations of DKOs alphabetically.

    perturbed_2: pertubations at gene level.
    
    perturbed: same with perturbed_2, except replacing ('CTRL2-CTRL1', 'CTRL1-CTRL1', 'hRosa26-CTRL1') with 
    'control', and ordering two perturbations of DKOs alphabetically.




```python
#Adjust names of pertubations in pop1.cells
pop1.cells['perturbed_1'] = pop1.cells['perturbed_1'].map(lambda x: 'RB1_3-CTRL1' if x =='CTRL1-RB1_3' else x )
pop1.cells['perturbed_2'] = pop1.cells['perturbed_2'].map(lambda x: 'RB1-CTRL1' if x =='CTRL1-RB1' else x )
control_targets = ['CTRL2-CTRL1', 'CTRL1-CTRL1', 'hRosa26-CTRL1']
#control_targets = ['CTRL2-CTRL1', 'hRosa26-CTRL1']
pop1.cells['perturbed'] = pop1.cells['perturbed_2'].map(lambda x: 'CTRL-CTRL' if x in control_targets else x)

pop1.cells['perturbed_1_Ab.ordered'] = pop1.cells['perturbed_1'].map(lambda x: '-'.join(sorted(x.split('-'))) if 'CTRL' not in x   else x)
pop1.cells['perturbed'] = pop1.cells['perturbed'].map(lambda x: '-'.join(sorted(x.split('-'))) if 'CTRL' not in x   else x)

print(len(pop1.cells['perturbed_1_Ab.ordered'].unique()))
print(len(pop1.cells['perturbed'].unique()))

##perturbed_1_Ab.ordered is used for mapping sgRNA-level fitness
perturbed_CROPseq = pop1.cells['perturbed_1_Ab.ordered'].unique()
perturbed_CROPseq
```

**2. Mapping fitness data from CRISPR screening to pop2.cells**

```python
#Adjust names of pertubations in pop2.cells
pop2.cells['perturbed_1'] = pop2.cells['perturbed_1'].map(lambda x: 'RB1_3-CTRL1' if x =='CTRL1-RB1_3' else x )
pop2.cells['perturbed_2'] = pop2.cells['perturbed_2'].map(lambda x: 'RB1-CTRL1' if x =='CTRL1-RB1' else x )
control_targets = ['CTRL2-CTRL1', 'CTRL1-CTRL1', 'hRosa26-CTRL1']
#control_targets = ['CTRL2-CTRL1', 'hRosa26-CTRL1']
pop2.cells['perturbed'] = pop2.cells['perturbed_2'].map(lambda x: 'CTRL-CTRL' if x in control_targets else x)

pop2.cells['perturbed_1_Ab.ordered'] = pop2.cells['perturbed_1'].map(lambda x: '-'.join(sorted(x.split('-'))) if 'CTRL' not in x   else x)
pop2.cells['perturbed'] = pop2.cells['perturbed'].map(lambda x: '-'.join(sorted(x.split('-'))) if 'CTRL' not in x   else x)

print(len(pop2.cells['perturbed_1_Ab.ordered'].unique()))
print(len(pop2.cells['perturbed'].unique()))

##perturbed_1_Ab.ordered is used for mapping sgRNA-level fitness
perturbed_CROPseq = pop2.cells['perturbed_1_Ab.ordered'].unique()
perturbed_CROPseq
```

```python
#Load sgRNA_level fitness data
fit1 = pd.read_csv('./sequencing/outs_S2/outs/full.1.log2RPM_fitlm_20200224.csv', index_col=0)
fit2 = pd.read_csv('./sequencing/outs_S2/outs/full.2.log2RPM_fitlm_20200224.csv', index_col=0)
print("Shape - fitness replicates 1:", fit1.shape)
print("Shape - fitness replicates 2:", fit2.shape)

fit1.rename({'x': 'fitness1'}, axis=1, inplace=True)
fit2.rename({'x': 'fitness2'}, axis=1, inplace=True)

#Calculate average fitness from two replicate experiments
fit = pd.merge(fit1[['fitness1']], fit2[['fitness2']], left_index = True, right_index = True)
fit['fitness'] = fit.mean(axis = 1, skipna = True)
fit.reset_index(inplace = True)
fit['index'] = fit['index'].map(lambda x: x.replace('CTRL000', 'CTRL'))
fit.set_index('index', inplace = True)
print("Shape - fitness replicates 2:", fit.shape)
fit.sort_values('fitness', ascending = False)

#Mapping average sgRNA-level fitness to pop2.cells according the names of guides
pop2.cells = pd.merge(fit, pop2.cells, left_index = True, right_on = 'perturbed_1_Ab.ordered', how = 'right')
print(len(pop2.cells['perturbed_1_Ab.ordered'].unique()))
perturbed_found = pop2.cells['perturbed_1_Ab.ordered'].unique()
perturbed_found
#np.setdiff1d(perturbed_CROPseq, perturbed_found)
pop2.cells.sort_values('fitness', ascending = False)
pop2.cells
```
